```python
# Input/Output paths
frames_root = "/content/4"  # Root folder containing 158 subfolders (each with its own frames)
output_root = "/content/output_videos"  # Folder to save output videos

script_root = "/content/sh_files"  # Folder to store generated .sh scripts
frames_acb_root = "/content/amplified_acb_frames"  # Folder to save  output acb frames
amplified_root = "/content/amplified_videos"  # Folder to save amplified videos
amplified_frames_root = "/content/amplified_frames"  # Folder to save amplified frames

os.makedirs(output_root, exist_ok=True)
os.makedirs(amplified_root, exist_ok=True)
os.makedirs(frames_acb_root, exist_ok=True)
os.makedirs(script_root, exist_ok=True)
os.makedirs(amplified_frames_root, exist_ok=True)
```

I/P Frame to Video

```python
# Process each file_name (158 folders)
file_names = sorted(os.listdir(frames_root))

for file_name in file_names:
    frame_folder = os.path.join(frames_root, file_name)
    output_video_path = os.path.join(output_root, f"{file_name}.avi")

    # Get sorted frames
    frames = sorted([img for img in os.listdir(frame_folder) if img.endswith((".png", ".jpg", ".jpeg"))])
    if not frames:
        continue

    # Read first frame for dimensions
    first_frame = cv2.imread(os.path.join(frame_folder, frames[0]))
    height, width, _ = first_frame.shape

    # Define video writer
    fourcc = cv2.VideoWriter_fourcc(*"XVID")
    out = cv2.VideoWriter(output_video_path, fourcc, 5, (width, height))

    # Write frames to video
    for frame in frames:
        img = cv2.imread(os.path.join(frame_folder, frame))
        out.write(img)
    out.release()
    print(f"Video saved: {output_video_path}")
```

I/P Video to acb_frames

```python
# Input/Output paths already defined above
frames_root = "/content/1"  # Root folder containing 158 subfolders (each with its own frames)
output_root = "/content/output_videos"  # Folder to save output videos

script_root = "/content/sh_files"  # Folder to store generated .sh scripts
frames_acb_root = "/content/amplified_acb_frames"  # Folder to save  output acb frames
amplified_root = "/content/amplified_videos"  # Folder to save amplified videos
amplified_frames_root = "/content/amplified_frames"  # Folder to save amplified frames
```

```python
# List all video files in the output_root directory
video_files = [f for f in os.listdir(output_root) if f.endswith(('.avi', '.mp4', '.mkv'))]

# Generate shell scripts
for file_to_be_maged in video_files:
    video_name, video_format = os.path.splitext(file_to_be_maged)
    video_path = os.path.join(output_root, file_to_be_maged)
    acb_output_dir = os.path.join(frames_acb_root, video_name)

    sh_file_content = f'''VIDEO_NAME={video_name}
VIDEO_FORMAT={video_format}

mkdir -p ${{VIDEO_NAME}}
ffmpeg -i "{video_path}" -f image2 ${{VIDEO_NAME}}/%06d.png
python make_frameACB.py ${{VIDEO_NAME}}
mkdir -p "{acb_output_dir}"
mv ${{VIDEO_NAME}} "{acb_output_dir}/{video_name}"
'''

    script_path = os.path.join(script_root, f"{video_name}_preproc.sh")
    with open(script_path, 'w') as file:
        file.write(sh_file_content)

    # Execute the script
    os.system(f"bash {script_path}")

print("Shell script processing complete. ACB files saved in", frames_acb_root)
```

acb_frames to **output_amplified_video**

```python
amplified_root = "/content/amplified_videos"  # Folder to save amplified videos
amplified_frames_root = "/content/amplified_frames"  # Folder to save amplified frames

# Amplify Videos
frames_root = frames_acb_root  # Use ACB frames as input

for file_name in file_names:
    video_path = os.path.join(frames_acb_root, file_name)  # Correcting the path
    if not os.path.exists(video_path):
        print(f"Skipping {file_name}: Path not found -> {video_path}")
        continue

    video_name = file_name
    video_format = ".avi"
    amplified_video_dir = os.path.join(amplified_root, video_name)
    os.makedirs(amplified_video_dir, exist_ok=True)

    amplified_frame_dir = os.path.join(amplified_frames_root, video_name)
    os.makedirs(amplified_frame_dir, exist_ok=True)

    # Load test data
    config.data_dir = video_path  # Fix: Correct input path
    data_loader = get_gen_ABC(config, mode=f'test_on_{video_name}')
    vid_size = cv2.imread(data_loader.paths[0]).shape[:2][::-1]

    # Amplify frames
    for amp in [10, 25, 50]:
        frames = []
        data_loader = get_gen_ABC(config, mode=f'test_on_{video_name}')

        for _ in range(data_loader.data_len):
            batch_A, batch_B = data_loader.gen_test()
            amp_factor = numpy2cuda(amp).unsqueeze(-1)
            with torch.no_grad():
                y_hats = model_test(batch_A, batch_B, 0, 0, amp_factor, mode='evaluate')

            for y_hat in y_hats:
                y_hat = unit_postprocessing(y_hat, vid_size=vid_size)
                frames.append(y_hat)

        # Save amplified video
        amp_video_path = os.path.join(amplified_video_dir, f"{video_name}_amp{amp}.avi")
        out = cv2.VideoWriter(amp_video_path, cv2.VideoWriter_fourcc(*'DIVX'), 5, frames[0].shape[-2::-1])
        for frame in frames:
            frame = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)
            cv2.putText(frame, f'amp_factor={amp}', (7, 37), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
            out.write(frame)
        out.release()
        print(f"Amplified video saved: {amp_video_path}")

        # Save amplified frames
        amp_frame_subdir = os.path.join(amplified_frame_dir, f"amp{amp}_frames")
        os.makedirs(amp_frame_subdir, exist_ok=True)
        for i, frame in enumerate(frames):
            cv2.imwrite(os.path.join(amp_frame_subdir, f"{i:06d}.png"), frame)
        print(f"Amplified frames saved in: {amp_frame_subdir}")
```
